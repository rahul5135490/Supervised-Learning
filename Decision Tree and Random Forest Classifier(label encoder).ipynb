```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
animals=pd.read_csv("horse.csv")
```

```python
animals.head()
```

```python
target=animals["outcome"]
```

```python
target.unique()
```

```python
animals=animals.drop(["outcome"], axis=1)
```

```python
animals.head()
```

```python
animals.info()
```

```python
category_variables=["surgery","age","temp_of_extremities", "peripheral_pulse","mucous_membrane","capillary_refill_time","pain","peristalsis","abdominal_distention","nasogastric_tube","nasogastric_reflux","rectal_exam_feces","abdomen","abdomo_appearance","surgical_lesion","cp_data"]
```

```python
for word in category_variables:
    animals = pd.get_dummies(animals)
```

```python
animals
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X=animals
```

```python
y=target
```

```python
X.head()
```

```python
y.unique()
```

```python
label_encoder=LabelEncoder()
```

```python
y= label_encoder.fit_transform(y)
```

```python
y
```

```python
 X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
print(X_train.shape)
```

```python
X_train.head()
```

```python
print(X_test.shape)
```

```python
X_test.head()
```

```python
# impute the missing value
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.NaN,strategy="most_frequent")
```

```python
X_train= imp.fit_transform(X_train)
```

```python
X_test= imp.fit_transform(X_test)
```

```python
X_train
```

```python
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred,y_test)
```

```python
print(accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier=RandomForestClassifier()
```

```python
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
accuracy=accuracy_score(y_pred,y_test)
print(accuracy)
```
